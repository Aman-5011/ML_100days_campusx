# ordinal encoding

```python
from sklearn.model_selection import train_test_split as tts
xtr,xts,ytr,yts=tts(df.drop('purchased' , axis=1) , df['purchased'] , test_size=0.2 , random_state=0)
```

```python
xtr.shape , xts.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
xtr=oe.fit_transform(xtr)
xts=oe.transform(xts)
```
