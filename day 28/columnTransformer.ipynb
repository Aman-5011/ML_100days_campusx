```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#fever par simpleimputer , gender and city par ohe and cough par ordinal encoding
```

```python
xtr,xts,ytr,yts=train_test_split(df.drop(columns=['has_covid']) , df['has_covid'],test_size=0.2)
```

# Without using column transformer

```python
si=SimpleImputer()
xtr_fever=si.fit_transform(xtr[['fever']])  # Double brackets keep it 2D; sklearn requires 2D input.
xts_fever=si.transform(xts[['fever']])
```

```python
oe=OrdinalEncoder(categories=[['Mild','Strong']])
xtr_cough=oe.fit_transform(xtr[['cough']])
xts_cough=oe.transform(xts[['cough']])
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False)
xtr_gender_city=ohe.fit_transform(xtr[['gender','city']])
xts_gender_city=ohe.transform(xts[['gender','city']])
```

```python
xtr_gender_city
```

```python
xtr_age=xtr[['age']].values  #values=It converts a DataFrame (or Series) into a NumPy array.
xts_age=xts[['age']].values
```

```python
xtr_transformed=np.concatenate((xtr_age,xtr_fever,xtr_gender_city,xtr_cough),axis=1)
xts_transformed=np.concatenate((xts_age,xts_fever,xts_gender_city,xts_cough),axis=1)
```

# now using column transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
trans=ColumnTransformer(transformers=[
    ('tnf1' , SimpleImputer() , ['fever']),
    ('tnf2' , OrdinalEncoder(categories=[['Mild','Strong']]) , ['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
] , remainder='passthrough'
)
```

```python
trans.fit_transform(xtr)
```

```python
trans.transform(xts)
```
