```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer , KNNImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('train (1).csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
xtr,xts,ytr,yts=train_test_split(df.drop(columns=['Survived']) , df['Survived'],test_size=0.2,random_state=2)
```

```python
knn=KNNImputer(n_neighbors=1,weights='distance')
xtr_trf=knn.fit_transform(xtr)
xts_trf=knn.transform(xts)
```

```python
lr=LogisticRegression()
lr.fit(xtr_trf,ytr)
y_pred=lr.predict(xts_trf)
accuracy_score(yts,y_pred)
```

```python
si=SimpleImputer()
xtr_trf2=si.fit_transform(xtr)
xts_trf2=si.transform(xts)
```

```python
lr=LogisticRegression()
lr.fit(xtr_trf2,ytr)
y_pred2=lr.predict(xts_trf2)
accuracy_score(yts,y_pred2)
```
